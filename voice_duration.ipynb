```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make sure you have seaborn version 0.11.0 or newer, as older versions may not support some features
print(sns.__version__)

# Replace 'voice duration.csv' with the full path to your CSV file if it's not in the current working directory
csv_file_path = '/Users/ruiqingfan/Documents/INDE-577-24/INDE-577-24-1/In-class notes/Datasets/voice duration.csv'

# Load the data into a DataFrame
df = pd.read_csv(csv_file_path)

# Plotting
plt.figure(figsize=(15, 10))

# Use seaborn's lmplot to create a regression line for each patient
# Setting ci=None will remove the confidence interval shading
sns.lmplot(data=df, x='Day', y='duration_min_adjusted', hue='Patient', markers='o',
           aspect=2, height=6, ci=None, palette='bright', legend=True)

# Adjust the legend outside of the plot area
#plt.legend(bbox_to_anchor=(1.05, 2), loc=1, borderaxespad=0.5)

# Set the title and labels
plt.title('Duration of Activity by Day for Each Patient')
plt.xlabel('Day')
plt.ylabel('Adjusted Duration (min)')

# Show the plot
plt.show()
```
